```python
import pandas as pd

file_path = r'C:\Users\GhidiniLucas\Desktop\Desktop\Projetos\Projeto AduaneiroETL\data\dados_aduaneiros.csv'

def load_data(file_path):
    try:
        data = pd.read_csv(file_path, sep = ';')
        print("Dados carregados com Sucesso.")
        return data
    except Exception as e:
        print(f'Erro ao carregar os dados: {e}')
    return None
```

```python
data = load_data(file_path)
data.head()
```
